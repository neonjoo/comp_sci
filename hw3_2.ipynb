## Q1

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 1 - np.exp(-x)

N = 2000
xx = []
yy = []
underx = np.array([])
undery = np.array([])
for i in range(N):
    x = np.random.uniform(0, 1)
    y = np.random.uniform(0, 1 - np.exp(-1))
    xx.append(x)
    yy.append(y)
    if y < f(x):
        underx = np.append(underx, x)
        undery = np.append(undery, y)

plt.figure()
plt.plot(xx, yy, '.')
plt.plot(underx, undery,'.')
plt.plot(np.linspace(0,1,50), f(np.linspace(0,1,50)))
plt.show()
```

## Q2

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 1 - np.exp(-x)

N = 3000
xx = []
yy = []
yybad = []
xxbad = []
underx = np.array([])
undery = np.array([])
for i in range(N):
    x = np.random.uniform(0, 1)
    ybad = np.random.uniform(0, 2*x)
    xxbad.append(x)
    x = np.sqrt(x)
    y = np.random.uniform(0, 2*x)
    
    xx.append(x)
    yy.append(y)
    yybad.append(ybad)
    if y < f(x):
        underx = np.append(underx, x)
        undery = np.append(undery, y)

plt.figure()
plt.plot(xx, yy, '.')
plt.plot(underx, undery,'.')
plt.plot(np.linspace(0,1,50), f(np.linspace(0,1,50)))
plt.plot(np.linspace(0,1,50), np.linspace(0,1,50))
plt.show()

plt.figure()
plt.plot(xxbad, yybad, 'b.')
#plt.plot(underx, undery,'.b')
plt.plot(np.linspace(0,1,50), f(np.linspace(0,1,50)))
plt.show()

plt.figure()
plt.plot(xx, yy, 'b.')
plt.plot(np.linspace(0,1,50), f(np.linspace(0,1,50)))
plt.plot(underx, undery,'b.')
plt.show()

```
